```python
import numpy as np

import matplotlib.pyplot as plt # for plotting
import matplotlib
matplotlib.rcParams['figure.dpi']=300 # highres display
```

Simple case of representation in 2D (no layer)

```python
def apply_net(y_in):  #return f(z) from y_in
    global w,b
    z=np.dot(w,y_in)+b
    return(1/(1+np.exp(-z)))
```

```python
#initiatlisation
N0=2
N1=1
w=np.random.uniform(low=-10,high=10,size=(N1,N0))
b=np.random.uniform(low=-1,high=1,size=N1)
```

```python
apply_net([0.8,0.3])
```

```python
#NOT EFFICIENT
#REPRESENTATION IN 2D
#utiliser : np.meshgrid(np.linspace(-0.5,0.5,M)

M=50
y_out=np.zeros([M,M])

for j1 in range(M):
    for j2 in range(M):
        value1=j1/M-0.5
        value2=j2/M-0.5
        y_out[j1,j2]=apply_net([value1,value2])[0]
```

```python
plt.imshow(y_out,origin='lower',extent=(-0.5,0.5,-0.5,0.5))
plt.colorbar()
plt.title("NN output as a function of input values")
plt.xlabel("y_2")
plt.ylabel("y_1")
plt.show()
```

```python
def apply_layer(y_in,w,b):
    z=np.dot(w,y_in)+b
    return(1/(1+np.exp(-z)))
```

```python
def apply_net(y_in):
    global w1,b1,w2,b2
    y1=apply_layer(y_in,w1,b1)
    y2=apply_layer(y1,w2,b2)
    return(y2)
```

```python
N0=2
N1=1000
N2=1
w1=np.random.uniform(low=-10,high=10,size=(N1,N0))
w2=np.random.uniform(low=-10,high=10,size=(N2,N1))
b1=np.random.uniform(low=-1,high=1,size=N1)
b2=np.random.uniform(low=-1,high=1,size=N2)
```

```python
M=1000
y_out=np.zeros([M,M])

for j1 in range(M):
    for j2 in range(M):
        value1=j1/M-0.5
        value2=j2/M-0.5
        y_out[j1,j2]=apply_net([value1,value2])[0]
```

```python
# display image
plt.imshow(y_out,origin='lower',extent=(-0.5,0.5,-0.5,0.5))
plt.colorbar()
plt.title("NN output as a function of input values")
plt.xlabel("y_2")
plt.ylabel("y_1")
plt.show()
```
